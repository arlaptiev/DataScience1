```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()  # plot styling
import pandas as pd
import numpy as np
```

```python
#setting data
df = pd.read_csv('salaries.csv', sep=',',header=None, skiprows=1)
df.drop(df.columns[[0, 4, 7]], axis=1, inplace=True)
data = np.array(df)

dic = {'Engineering':'0','Party':'1','Liberal Arts':'2','Ivy League':'3','State':'4'}
for row in np.arange(len(data)):
    data[row][0] = dic[data[row][0]]
    for el in np.arange(len(data[row])):
        data[row][el] = float(data[row][el].replace('$','').replace(',',''))
        
X = data[:, 1:5]
y = data[:, 0]
plt.scatter(X[:, 0], X[:, 1], s=50);
y
```

```python
#predict using sklearn
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=5)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)
labels = kmeans.labels_
y_kmeans
```

```python
#plot prediction
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')

centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5);
```
