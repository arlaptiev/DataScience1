```python
import random
from codecs import open
from textblob.classifiers import NaiveBayesClassifier
import numpy as np
import collections
```

```python
def read_labeled_data(path):
    with open(path, encoding='utf8') as f:
        X = []
        y = []
        for line in f:
            target, _, sentence = line.partition(',')
            X.append(sentence)
            y.append(int(target))
    return X, y
```

```python
def get_model_predictions():
    model = Model()
    model.fit(*read_labeled_data("./data/train.txt"))
    with open('data/test.txt', encoding='utf8') as f:
        for line in f:
            yield line, model.predict(line)
            
def score(predictions):
    y = open("./data/dev.y").read().replace('\n', '')
    y = str(bin(int(y, 16))[2:])
    y_pred = [str(p) for _, p in predictions[-len(y):]]
    correct = sum([y1 == y2 for y1, y2 in zip(y, y_pred)])
    accuracy = float(correct) / len(y)
    print("-" * 80)
    print(" " * 40 + "Accuracy: {:.3%}".format(accuracy))
```

```python
predictions = get_model_predictions()
predictions = list(predictions)
score(predictions)
```
